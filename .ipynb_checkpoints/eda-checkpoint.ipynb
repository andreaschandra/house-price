## Input Missing Values

```python
dtrain.PoolQC.fillna('None', inplace=True)

dtrain.MiscFeature.fillna('None', inplace=True)

dtrain.Alley.fillna('None', inplace=True)

dtrain.Fence.fillna('None', inplace=True)

dtrain.FireplaceQu.fillna('None', inplace=True)

dtrain.LotFrontage.fillna(dtrain.LotFrontage.mean(), inplace=True)

dtrain.GarageCond.fillna('None', inplace=True)

dtrain.GarageType.fillna('None', inplace=True)

dtrain.GarageFinish.fillna('None', inplace=True)

dtrain.GarageQual.fillna('None', inplace=True)

dtrain.BsmtFinType2.fillna('None', inplace=True)

dtrain.BsmtExposure.fillna('None', inplace=True)

dtrain.BsmtQual.fillna('None', inplace=True)

dtrain.BsmtCond.fillna('None', inplace=True)

dtrain.BsmtFinType1.fillna('None', inplace=True)

dtrain.MasVnrType.fillna('None', inplace=True)

dtrain.Electrical.fillna('None', inplace=True)
```

## EDA

```python
plt.figure(figsize = (12, 8))
sns.distplot(dtrain.SalePrice)
```

```python
plt.figure(figsize = (20, 16))
sns.heatmap(dtrain[col_numeric].corr(), annot = True)
```

Histogram of numeric variable

```python
fig = plt.figure(figsize=(20, 18))
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1, len(col_numeric)):
    ax = fig.add_subplot(5, 7, i)
#     ax.hist(dtrain[col_numeric[i]])
    sns.distplot(dtrain[col_numeric[i]])
    plt.xlabel(col_numeric[i])
```

scatter plot variable numerik dengan sales price

```python
fig = plt.figure(figsize=(20, 40))
fig.subplots_adjust()
# fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1, len(col_numeric[:-1])):
    axs = fig.add_subplot(8, 4, i)
#     plt.scatter(x = col_numeric[i], y = col_numeric[-1], data = dtrain)
    sns.scatterplot(x = col_numeric[i], y = col_numeric[-1], data = dtrain)
    plt.xlabel(col_numeric[i])
    plt.ylabel(col_numeric[-1])
```

Category frequency

```python
def make_frequency(index):
    vc = dtrain[col_category[index]].value_counts().reset_index()
    vc.columns = ['category', 'frequency']
    
    return vc
```

```python
fig = plt.figure(figsize=(20, 38))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

for i in range(1, len(col_category)+1):
    ax = fig.add_subplot(10, 5, i)
    vc = make_frequency(i-1)
    sns.barplot(x = 'category', y = 'frequency', data = vc)
    plt.xlabel(col_category[i-1])
    
plt.savefig('category_frequency.jpg')
```
